### Normes
On définit les deux fonctions suivantes, destinées à prendre en entrée des vecteurs `x` et `y`.

```python
def f1(x, y):
    return np.square(np.abs(np.subtract(x, y)))

def f2(x, y):
    return np.square(np.linalg.norm(np.subtract(x, y)))
```

```python
# Vos tests ici
x = [3, 2, 4]
y = [-2, 3, 1]

print(f1(x, y))
print(f2(x, y))

x = 3
y = -2

print(f1(x, y))
print(f2(x, y))

# Pour la Question 2 : version plus efficace
z = np.subtract(x, y)
print(np.inner(z, z))

# Pour la Question 2 : effectuer des opérations en trop peut augmenter les chances d'erreurs numériques
print(np.sqrt(10000000000000) ** 2)
```

### Question 1
- Que calculent ces deux fonctions ? Expliquez la différence entre les deux.
- Si `x` et `y` sont des vecteurs à $n$ éléments, quelle est la taille des vecteurs retournés ? Pourquoi ?
- Qu'observez-vous quand `x` et `y` ont taille $1$ ?
- Dans la fonction de poids $w^{(i)}$ de la régression locale, laquelle est à utiliser ?

*Réponses :* La fonction `f1` contient trois opérations qui s'effectuent composante par composante: ce qui est retourné est un vecteur $z\in\mathbb{R}^n$ tel que $z_i = |x_i - y_i|^2$ (et où $|\cdot|$ est la valeur absolue, qui est une fonction $\mathbb{R} \to \mathbb{R}$). La fonction `f2` utilise la norme euclidienne, qui est une opération qui calcule un scalaire à partir d'un vecteur ; `f2` retourne $||x - y||^2\in\mathbb{R}$ (où $||\cdot||$ est la norme euclidienne, qui est une fonction $\mathbb{R}^n \to \mathbb{R}$).

Lorsque $x$ et $y$ ont taille $1$, les deux fonctions coïncident, car la norme euclidienne à une dimension correspond à la valeur absolue.

La fonction de poids $w^{(i)}$ est un scalaire, donc la fonction `f1` n'est pas adaptée. Ici, la fonction `f2` correspond au numérateur dans l'exposant de $w^{(i)}$.

### Question 2
Supposons maintenant que, pour un vecteur $x$, vous vouliez calculer $||x||^2$, où $||\cdot||$ est la norme euclidienne. Comment le faire de façon efficace (avec ou sans `numpy`) ? Expliquez pourquoi `np.linalg.norm(x) ** 2` effectue des calculs inutiles.

*Réponse :* La fonction `np.linalg.norm(x)` retourne $\sqrt{\sum_{i = 1}^n x_i^2}$ ; quand on la prend au carré, on calcule juste $\sum_{i = 1}^n x_i^2$. Quand on appelle `np.linalg.norm(x) ** 2`, l'ordinateur calcule donc une racine carrée, et prend le carré tout de suite après. L'opération racine carrée peut être relativement coûteuse quand elle est appelée fréquemment, mais n'est ici pas utile ; il est plus efficace de calculer directement $\sum_{i = 1}^n x_i^2$. C'est ce que calcule le produit scalaire `np.inner(x, x)`. Éviter d'effectuer l'opération $\sqrt{\cdot}^2$ diminue aussi les risques d'erreurs numériques.

### Matrices
On définit les trois fonctions suivantes, destinées à prendre en entrée une matrice `X` à deux dimensions.

```python
def g1(X):
    return np.linalg.inv(X)

def g2(X):
    return 1 / X

def g3(X):
    return np.power(X, -1)
```

```python
# Vos tests ici
X = np.array([[3., 2.], [4., 5.]])

print(g1(X))
print(g2(X))
print(g3(X))

# On vérifie que la matrice retournée par g1 satisfait bien la définition de l'inverse (à une légère erreur numérique près)
print(X @ g1(X))
print(g1(X) @ X)

# Ce n'est pas du tout le cas pour g2 et g3 : le produit de X avec g2(X) n'a pas de sens particulier
print(X @ g2(X))
```

### Question 3
- Que calculent ces trois fonctions ? Expliquez leurs différences.
- Dans la formule exacte pour la régression linéaire (qui s'écrit mathématiquement $(X^TX)^{-1}X^TY$), laquelle de ces formules allez-vous utiliser ? Pourquoi ?

*Réponses :* Les fonctions `g2` et `g3` effectuent la même opération : simplement, prendre l'inverse de chaque élément de la matrice, composante par composante. Par contre, `g1` calcule la matrice inverse de la matrice $X$, c'est-à-dire l'opération notée $X^{-1}$ lors du cours. La matrice inverse $X^{-1}$ d'une matrice $X$ (quand elle existe) est la seule matrice telle que $XX^{-1}$ et $X^{-1}X$ sont égales à la matrice identité. Les opérations effectuées par `g2` et `g3` ont moins de propriétés mathématiques utiles et ne devraient pas être utilisées dans le cours.

On définit les trois fonctions suivantes, destinées à prendre en entrée des vecteurs `x` et `y`.

```python
def h1(x, y):
    return x.T @ y

def h2(x, y):
    x = x[:,None]
    y = y[:,None]
    return x.T @ y

def h3(x, y):
    x = x[:,None]
    y = y[:,None]
    return x @ y.T
```

```python
# Vos tests ici
x = np.array([3, 2, 4])
y = np.array([-2, 3, 1])
print(x[:,None])

print()
print(h1(x, y))
print(h2(x, y))
print(h3(x, y))

print()
print(np.inner(x, y))
print(np.outer(x, y))
```

### Question 4
Que calculent ces trois fonctions ? Expliquez pourquoi elles donnent un résultat différent.

*Réponses :* Les fonctions `h1` et `h2` calculent toutes les deux d'une certaine façon le produit scalaire de $x$ et $y$. Pour `h1`, $x$ et $y$ sont deux vecteurs à une dimension, et numpy retourne leur produit scalaire. Pour `h2`, $x$ et $y$ sont des vecteurs avec une deuxième dimension, et numpy effectue un produit matriciel entre une matrice $1\times n$ et une matrice $n \times 1$, et retourne donc une matrice $1 \times 1$. Celle-ci contient comme seule entrée le produit scalaire de $x$ et $y$. Notez que l'opération effectuée par `h1` est la même que celle effectuée par l'appel `np.inner(x, y)`.

L'opération effectuée par `h3` retourne un résultat très différent : c'est une multiplication matricielle (car les deux vecteurs ont deux dimensions) entre un vecteur de taille $n\times 1$ et un vecteur de taille $1\times n$, ce qui retourne une matrice $n \times n$. Celle-ci contient comme élément $(i, j)$ la valeur $x_i * y_j$. Cette opération correspond à la fonction numpy `np.outer(x, y)`, que certains d'entre vous ont utilisée au TP2.

## Méthode de Newton

On considère la fonction $f\colon \mathbb{R}^2 \to \mathbb{R}$ telle que $$f(x, y) = (x^2 + y - 11)^2 + (x + y^2 - 7)^2.$$

```python
import matplotlib.pyplot as plt
from matplotlib import cm

def f(x, y):
    return (x ** 2 + y - 11) ** 2 + (x + y ** 2 - 7) ** 2

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

X = np.arange(-5, 5, .1)
Y = np.arange(-5, 5, .1)
X, Y = np.meshgrid(X, Y)
f_vect = np.vectorize(f)
Z = f_vect(X, Y)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, alpha=.7,
                       linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=.5, aspect=5)
plt.show()
```

1. En utilisant la méthode de Newton à plusieurs dimensions, on souhaite trouver les minimums et maximums locaux de $f$. Commencez par calculer le gradient et la matrice hessienne de $f$, et implémentez la méthode de Newton.

*Réponse :*
$$\nabla f = (4x^3 + 4xy - 42x + 2y^2 - 14, 2x^2 - 26y - 22 + 4xy + 4y^3)$$
$$H_f = \left(\begin{matrix}
             12x^2 + 4y - 42     &4x + 4y \\
             4x + 4y             &4x + 12y^2 - 26
        \end{matrix}\right)$$

```python
def gradient_f(x, y):
    dx = 4*x**3 + 4*x*y - 42*x + 2*y**2 - 14
    dy = 2*x**2 + 4*x*y - 26*y + 4*y**3 - 22
    return np.array([dx, dy])

def hessian_f(x, y):
    dxx = 12*x**2 + 4*y - 42
    dxy = 4*x + 4*y
    dyy = 4*x + 12*y**2 - 26
    return np.array([[dxx, dxy], [dxy, dyy]])

def newton_multidim(f, df, theta, max_iter=1000, epsilon=1e-10):
    """Finds a root of multivariate function f given its gradient df and an initial guess theta.
       Assumes that a value theta is a root if np.linalg.norm(f(theta)) < epsilon.
       Raises an exception if no such value is found within max_iter iterations."""
    f_theta = f(*theta)
    i = 0
    while np.linalg.norm(f_theta) > epsilon and i < max_iter:
        theta = theta - np.linalg.inv(df(*theta)) @ f_theta
        f_theta = f(*theta)
        i += 1
    if np.linalg.norm(f_theta) > epsilon and i >= max_iter:
        raise Exception("Root of f not found within max_iter iterations in newton_multidim.")
    return theta
```

2. Implémentez une façon de trouver plusieurs extremums en faisant démarrer la méthode de Newton de nombreux points de l'espace. Faites attention à ne pas inclure plusieurs fois les mêmes extremums avec des valeurs numériques légèrement différentes. Pour cela, implémentez la méthode `add_point_if_far` qui ajoute un point à une liste si et seulement si le point n'est à distance $< \epsilon$ d'aucun autre point dans la liste.

Note : il est difficile d'être sûr d'avoir trouvé exhaustivement tous les extremums avec des méthodes numériques ; cette fonction a au moins 9 points où le gradient s'annule, incluant au moins quatre minimums locaux et un minimum local.

```python
def add_point_if_far(points, new_point, eps=1e-6):
    """Given a list `points` of points of the same dimension, adds the point `new_point` to the list
       if the distance from new_point to any point of the list `points` is not smaller than `eps`."""
    if all(np.linalg.norm(np.subtract(point, new_point)) > eps for point in points):
        points.append(new_point)

# Find all (or at least a few) points where the gradient vanish.
# For future reference, it is "Himmelblau's function". Multiple sources discuss it online.
roots = []
X = np.arange(-5, 5, .2)
Y = np.arange(-5, 5, .2)
for x in X:
    for y in Y:
        root = newton_multidim(gradient_f, hessian_f, [x, y])
        add_point_if_far(roots, root)

roots = np.array(roots)
print(roots)
```

## Régression locale avec descente de gradient

Nous allons ici retravailler la régression locale à une variable. L'énoncé sera le même qu'au TP2, mais on vous demande d'utiliser la descente de gradient plutôt que la méthode exacte pour trouver les valeurs optimales de $\theta$.

On considère un nouveau dataset, fourni dans `tp5.zip` sous le nom `tips.npy` (repris de [Kaggle](https://www.kaggle.com/code/yashsharmabharatpur/tips-collected-at-a-restaurant)). Celui-ci contient $244$ valeurs de pourboires donnés à des serveurs en fonction du prix de l'addition. Toutes les valeurs sont en $.

```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.load("tips.npy") # 244 données, deux dimensions : prix de l'addition, pourboire reçu
print(np.shape(dataset))
print(dataset[:10]) # On affiche les 10 premiers

bills = dataset[:,0]
tips = dataset[:,1]
```

```python
plt.figure()
plt.plot(bills, tips, "o")
plt.xlabel("Bills")
plt.ylabel("Tips")
plt.grid()
plt.show()
```

Réimplémentez et affichez les prédictions données par une **régression linéaire locale** en utilisant la descente de gradient :
- implémentez la fonction de poids,
- choisissez une valeur de bande passante adaptée aux données,
- implémentez la fonction qui calcule les prédictions, en utilisant une descente de gradient,
- affichez vos prédictions sur un graphique, en superposant la courbe aux données.

```python
# Fonction de poids

def weight(x, xi, tau):
    """Returns the weight to give xi from the training set
       with respect to a point x and with bandwidth tau."""
    diff = np.subtract(xi, x)
    return np.exp(- np.inner(diff, diff) / (2 * (tau ** 2)))

def weight_vector(x, X, tau):
    """Returns the vector of weights for each element of vector X
       with respect to a point x and with bandwidth tau."""
    return np.array([weight(x, xi, tau) for xi in X])
```

```python
# Graphique des poids

plt.figure()
x = 20
plt.plot(bills, weight_vector(x, bills, 1), "o", label="tau=1")
plt.plot(bills, weight_vector(x, bills, 3), "o", label="tau=3") # 3 seems to be a good value
plt.plot(bills, weight_vector(x, bills, 20), "o", label="tau=20")
plt.legend()
plt.xlabel("Bills")
plt.ylabel("Weight")
plt.show()
```

```python
# Calcul des matrices X et Y

X = bills[:,None]
m = X.shape[0]
X = np.concatenate((np.ones((m, 1)), X), axis=1)
Y = tips
```

```python
# Descente de gradient et fonction de prédiction.

# On utilise ici la descente de gradient, comme au TP1 mais en ajoutant la matrice W dans la formule.

# On passe des paramètres avec des produits déjà calculés, car les mêmes produits reviennent à chaque itération.
def gradient_square_error(Theta, XTWX, XTWY):
    """Returns the gradient of the square error on training examples (X, Y) with weight matrix W.
       The result is a matrix with the same dimension as Theta."""
    return 2 * (XTWX @ Theta - XTWY)

def one_step_linear_regression_gradient_descent(Theta, XTWX, XTWY, learning_rate):
    """One step of the gradient descent for the multiple linear regression.
       Takes the parameters, the training examples and the learning rate, and returns the new values for the parameters."""
    return Theta - learning_rate*gradient_square_error(Theta, XTWX, XTWY)

def theta_minimizing_error(X, Y, x, tau, nb_steps, learning_rate = 0.01, weight_function=weight_vector):
    m, n = X.shape

    # Construction de la matrice W
    weights = weight_function(x, X, tau)
    W = np.zeros((m, m))
    for i in range(m):
        W[i][i] = .5 * weights[i]
    
    # Précalculs pour éviter des multiplications matricielles redondantes
    XTW = X.T @ W
    XTWX = XTW @ X
    XTWY = XTW @ Y

    Theta = np.zeros((n))
    for _ in range(nb_steps):
        Theta = one_step_linear_regression_gradient_descent(Theta, XTWX, XTWY, learning_rate / m)

    return Theta

def prediction(X, Y, x, tau, weight_function=weight_vector):
    one_x = np.concatenate(([1], np.atleast_1d(x)))
    theta = theta_minimizing_error(X, Y, one_x, tau, 5000, weight_function=weight_function)
    return one_x.T @ theta
```

```python
# We compare with the exact method to be sure (this was not needed in the TP)
def theta_minimizing_error_exact(X, Y, x, tau, weight_function=weight_vector):
    m, _ = X.shape

    # Construction de la matrice W
    weights = weight_function(x, X, tau)
    W = np.zeros((m, m))
    for i in range(m):
        W[i][i] = .5 * weights[i]
    
    # Précalcul pour éviter une multiplication matricielle redondante
    XTW = X.T @ W

    # Formule exacte vue au cours
    return np.linalg.inv(XTW @ X) @ XTW @ Y

def prediction_exact(X, Y, x, tau, weight_function=weight_vector):
    one_x = np.concatenate(([1], np.atleast_1d(x)))
    theta = theta_minimizing_error_exact(X, Y, one_x, tau, weight_function=weight_function)
    return one_x.T @ theta
```

```python
# Affichage des résultats en plus des données.

nb_points = 300
to_evaluate = np.linspace(np.min(X[:,1]), np.max(X[:,1]), nb_points)
plt.figure()
plt.plot(X[:,1], Y, 'o')
plt.xlabel("Bills")
plt.ylabel("Tips")

tau = 3
predictions = [prediction(X, Y, x, tau) for x in to_evaluate]
plt.plot(to_evaluate, predictions, '-', label="gradient descent")
predictions_exact = [prediction_exact(X, Y, x, tau) for x in to_evaluate] # Exact method as well to be sure
plt.plot(to_evaluate, predictions_exact, '-', label="exact formula")
plt.legend()
plt.show()
```
