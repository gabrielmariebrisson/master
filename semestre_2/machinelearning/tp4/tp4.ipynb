## 1. Gaussian Discriminant Analysis (GDA)

Nous considérons le jeu de données *IRIS*, qui est un dataset très classique (https://fr.wikipedia.org/wiki/Iris_de_Fisher). Celui-ci est inclus dans `scikit-learn`. Il contient 150 données sur 3 types d'iris (*Iris setosa*, *Iris virginica* et *Iris versicolor*). Les données ont 4 features: la longueur et la largeur des sépales, et la longueur et la largeur des pétales.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

dataset = load_iris()
# On divise notre dataset en des données d'entraînement et des données de test ; on ne va s'entraîner
# que sur les données d'entraînement (75 % du dataset), et on garde les données de test pour évaluer la qualité de nos prédictions.
x_train, x_test, y_train, y_test = train_test_split(dataset.data, dataset.target, train_size=0.75)
```

On affiche les données selon les deux premières features.

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(dataset.data[:, 0], dataset.data[:, 1], c=dataset.target)
ax.set(xlabel=dataset.feature_names[0], ylabel=dataset.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], dataset.target_names, loc="lower right", title="Classes"
)
```

On souhaite prédire le type d'iris en utilisant une analyse discriminante gaussienne. Pour cela, calculez les trois paramètres ($\mu$, $\Sigma$, et $\phi$) décrivant la gaussienne de chaque classe d'iris.

```python
def fit(x_train, y_train):
    m = y_train.shape[0] # Number of training examples.

    x_train = x_train.reshape(m, -1)
    input_feature = x_train.shape[1] # Number of input features. In our case, 4.
    class_label = len(np.unique(y_train.reshape(-1))) # Number of classes. In our case, 3.
    
    mu = np.zeros((class_label, input_feature))
    sigma = np.zeros((class_label, input_feature, input_feature))
    phi = np.zeros(class_label)

    for i in range(class_label):
        # Extract examples of the current class
        x_class = x_train[y_train == i]

        # Calculate mu for the current class
        mu[i] = np.mean(x_class, axis=0)

        # Calculate sigma for the current class
        sigma[i] = np.cov(x_class, rowvar=False)

        # Calculate phi for the current class
        phi[i] = len(x_class) / m
    
    return phi, mu, sigma
```

À partir des paramètres calculés avec la fonction `fit`, implémentez une fonction `predict` qui calcule la classe la plus probable pour chacune des données de test.

```python
from scipy.stats import multivariate_normal

def predict(x_tests, phi, mu, sigma):
    # flatten the training data
    x_tests = x_tests.reshape(x_tests.shape[0], -1)
    class_label = mu.shape[0] # Number of classes. In our case, k = 3.
    
    
    predictions = np.zeros(x_tests.shape[0], dtype=int)

    for i in range(x_tests.shape[0]):
        likelihoods = np.zeros(class_label)

        for j in range(class_label):
            # Calculate the likelihood of the current sample for each class
            likelihoods[j] = multivariate_normal.pdf(x_tests[i], mean=mu[j], cov=sigma[j])

        # Multiply by the prior probability (phi) for each class
        posterior_probabilities = likelihoods * phi

        # Predict the class with the highest posterior probability
        predictions[i] = np.argmax(posterior_probabilities)

    return predictions
```

Testez votre fonction `predict` sur les données de test. Pour évaluer la qualité de vos prédictions, calculez le score $\mathsf{F}_1$ de vos prédictions. Vous pouvez utiliser la méthode `f1_score` de `scikit-learn` (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html).

```python
from sklearn.metrics import f1_score

phi, mu, sigma = fit(x_train, y_train)

predictions = predict(x_test, phi, mu, sigma)

f1 = f1_score(y_test, predictions, average='weighted')

print("F1 Score:", f1)
```

Comparez vos prédictions avec la méthode `LinearDiscriminantAnalysis` de `scikit-learn` (https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html).
Il est possible que vos prédictions soient légèrement différentes.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda_model = LinearDiscriminantAnalysis()
lda_model.fit(x_train, y_train)

# Predict using the model
lda_predictions = lda_model.predict(x_test)

# Calculate F1 score for LinearDiscriminantAnalysis predictions
f1_lda = f1_score(y_test, lda_predictions, average='weighted')

print("F1 Score for LinearDiscriminantAnalysis:", f1_lda)
```

## 2. Spam filters avec Naive Bayes

```python
import nltk 
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

En utilisant `pandas`, on charge les données d'un dataset contenant des emails classifiés en spam ou non-spam (la valeur $1$ indique les spams, et la valeur $0$ les non-spams).

```python
df=pd.read_csv('spam_email_raw_text_for_NLP.csv')
df.head()
```

La colonne "FILE_NAME" n'est pas utile ; on la supprime.

```python
df.drop('FILE_NAME',axis=1,inplace=True)
df.head()
```

On compte le nombre de spams et de non-spams.

```python
df.CATEGORY.value_counts()
```

Les "stopwords" sont les petits mots qui apparaissent dans la plupart des textes et sont peu pertinents pour la classification en spam.

```python
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()
```

On crée le corpus. On le simplifie afin de retirer les caractères non-alphanumériques, les majuscules, et les caractères blancs. Le "lemmatizer" sert à identifier les mots similaires, par exemple "chiens" et "chien". Ce calcul peut prendre un peu de temps.

```python
from tqdm.notebook import tqdm

corpus=[]
for i in tqdm(range(len(df))):
    # removing all non-alphanumeric characters
    message = re.sub('[^a-zA-Z0-9]',' ',df['MESSAGE'][i]) 
    # converting the message to lowercase
    message = message.lower() 
    # spliting the sentence into words for lemmatization                 
    message = message.split()      
    # removing stopwords and lemmatizing            
    message = [lemmatizer.lemmatize(word) for word in message
             if word not in set(stopwords.words('english'))] 
    # Converting the words back into sentences
    message = ' '.join(message)    
    # Adding the preprocessed message to the corpus list            
    corpus.append(message)                 

print(corpus[:5])
```

Pour chaque email, on crée un vecteur booléen qui donne la présence ou l'absence de mot (ou séquences de mots) dans le message. On utilise la méthode `CountVectorizer` (https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html). Pour limiter la taille des vecteurs, on se limite aux 2500 mots les plus présents.

```python
cv = CountVectorizer(max_features = 2500, ngram_range = (1,3))
X = cv.fit_transform(corpus).toarray()
y = df['CATEGORY']
```

On sépare à nouveau les données en gardant 80 % de données d'entraînement et 20 % de données de test.

Grâce à la variable `cv`, on peut facilement transformer un email en vecteur booléen, et on peut voir à quels mots correspondent les 2500 features.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1, stratify=y)
print(cv.get_feature_names_out()[1100:1120]) # On affiche les mots correspondant à 20 features.
message = ["You won 10000 dollars, please provide your account details so that we can transfer the money"]
print(cv.transform(message))
print(cv.get_feature_names_out()[1596], cv.get_feature_names_out()[2317])
```

### 2.1 Naive Bayes à la main

Implémentez la méthode *naive Bayes* vue au cours pour prédire si un message donné est un spam ou non. Testez vos méthodes sur les données de test.

```python
def fit(x_train, y_train):
    
    y1 = np.sum(y_train == 1, axis=0)
    y0 = np.sum(y_train == 0, axis=0)
    
    x1_y1 = np.sum(x_train [y_train == 1], axis=0)
    x0_y0 = np.sum(x_train [y_train == 0], axis=0)
    
    phi_y1 = (x1_y1 + 1) / (y1 + 2)  
    phi_y0 = (x0_y0 + 1) / (y0 + 2)
    
    phi = y1 / y_train.shape[0]
    
    return phi, phi_y1, phi_y0
```

```python
def predict(x_test, phi, phi_y1, phi_y0):
    log_likelihood_y1 = np.sum(np.log(phi_y1) * x_test, axis=1)
    log_likelihood_y0 = np.sum(np.log(phi_y0) * x_test, axis=1)

    log_likelihood_ratio = log_likelihood_y1 - log_likelihood_y0

    log_odds_spam = np.log(phi) + log_likelihood_ratio

    log_odds_non_spam = np.log(1 - phi) + log_likelihood_ratio

    return  np.where(log_odds_spam > log_odds_non_spam, 1, 0)
```

```python
phi, phi_y1, phi_y0 = fit(x_train, y_train)

predictions = predict(x_test, phi, phi_y1, phi_y0)

f1 = f1_score(y_test, predictions, average='weighted')

print("F1 Score:", f1)
```

### 2.2 Naive Bayes avec Scikit-learn

Utilisez la méthode `MultinomialNB` de `scikit-learn` (https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html) pour effectuer la même tâche qu'en Section 2.1. Utilisez la méthode `classification_report` de `sklearn.metrics` pour évaluer et interpréter la qualité de vos prédictions.

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

clf = MultinomialNB()
clf.fit(x_train, y_train)

predictions = clf.predict(x_test)

report = classification_report(y_test, predictions)
print("Classification Report (No Laplace smoothing):\n", report)

clf_alpha = MultinomialNB()
clf_alpha.fit(x_train, y_train)
predictions_alpha = clf_alpha.predict(x_test)
report_alpha = classification_report(y_test, predictions_alpha)
print(f"Classification Report :\n", report_alpha)
```

Appliquez votre méthode à nouveau en augmentant le paramètre `alpha` correspondant au *Laplace smoothing*, et comparez les résultats.

```python
alpha = 0.0001
alpha_value=[]
precision = []
recall = []
f1 = []

while alpha <= 20:
    clf_alpha = MultinomialNB(alpha=alpha)
    clf_alpha.fit(x_train, y_train)
    predictions_alpha = clf_alpha.predict(x_test)
    report_alpha = classification_report(y_test, predictions_alpha, output_dict=True)

    precision += [report_alpha['1']['precision']]
    recall += [report_alpha['1']['recall']]
    f1 += [report_alpha['1']['f1-score']]
    
    alpha_value+=[alpha]
    alpha *= 1.5

def printGraphe(alpha_value,tab, label):
    plt.figure()
    plt.plot(alpha_value, tab, label=label)
    plt.xlabel('Alpha')
    plt.ylabel(label)
    plt.legend()
    
    plt.show()

printGraphe(alpha_value, precision, 'Precision')
printGraphe(alpha_value, recall, 'Recall')
printGraphe(alpha_value, f1, 'F1-Score')
```
